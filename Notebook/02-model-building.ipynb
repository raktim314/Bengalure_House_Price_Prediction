## Model Building

```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
DATA_PATH = "../data/processed/final_data.csv"
```

```python
df = pd.read_csv(DATA_PATH)
```

```python
df.head()
```

In the dataset above the feature is location is a categorical variable, so we need to encode this to buildig machine learning model.

```python
dummies = pd.get_dummies(df.location)
dummies.head()
```

```python
df2 = pd.concat([df.drop(columns=['location']), dummies.drop(columns=['other'])],axis='columns')
```

```python
df2.head()
```

```python
df2.shape
```

```python
features = df2.drop('price', axis='columns')
features.head()
```

```python
target = df2.price
target.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2,random_state=10)
```

```python
X_train
```

### Baseline Model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
cross_val_score(lr, features, target, cv=cv).mean()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
```

```python
def find_best_model_using_gridsearch(x,y):
    algos = {
        'linear_regression': {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['mse', 'friedman_mse'],
                'splitter': ['best','random']
            }
        }
        
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(x,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })
    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
```

```python
result = find_best_model_using_gridsearch(features, target)
```

## Predict Price

```python
def predict_price(location,sqft,bath,bhk):
    location_index = np.where(x.columns==location)[0][0]
    
    X = np.zeros(len(x.columns))
    X[0] = sqft
    X[1] = bath
    X[2] = bhk
    if location_index >= 0:
        X[location_index] = 1
        
    return lr.predict([X])[0]
```
